```python
# groupby(['match_id','striker'])
# Har match ke andar har batsman ke liye grouping.

# agg(runs_scored=("runs_off_bat", "sum"))
#  Wo batsman ne match me total runs kitne banaye.

# agg(balls_faced=("ball", "count"))
#  Usne kitni balls face ki.

# reset_index()
#  Result ko normal DataFrame banaya (nahi to multi-index hota).




batting = df.groupby(["match_id","striker"]).agg(run_scored = ("runs_off_bat","sum"),ball_faced = ("ball","count")).reset_index()
batting["strike_rate"] = (batting["run_scored"]/batting["ball_faced"]) * 100
batting.head()



```

```python
# groupby(['match_id','bowler'])
# Har match ke andar har bowler ke liye grouping.

# agg(runs_conceded=("runs_off_bat", "sum"))
#  Wo bowler ne match me total runs kitne diye.

# agg(balls_bowled=("ball", "count"))
#  Usne kitni balls fenki .

# wickets(('player_dismissed',"count"))
# kitne wicket liye

## economy rate count kiya 

# reset_index()
#  Result ko normal DataFrame banaya (nahi to multi-index hota).


bowling = df.groupby(["match_id","bowler"]).agg(run_conceded =("runs_off_bat","sum"),ball_bowled= ("ball","count"),wickets = ('player_dismissed',"count")).reset_index()
bowling["over"] = bowling["ball_bowled"]/6
bowling["economy_rate"] = bowling["run_conceded"]/bowling["over"]
bowling.head()





```

```python
# Rename before merging
# striker ,bowler ko player name diya (Matlab ek hi player kabhi batsman bhi hota hai, kabhi bowler bhi.)
#  merge bowling and batting on base of match_id and player
# how = outer(Agar koi player sirf batting karke gaya (bowling nahi ki), to uska bowling wala part NaN hota).
# .fillna(0) = put 0 in place of missing values
## example:
# Agar player ne bowling nahi ki → wickets=0, balls_bowled=0
# Agar batting nahi ki → runs_scored=0, balls_faced=0


batting.rename(columns={"striker": "player"}, inplace=True)
bowling.rename(columns={"bowler": "player"}, inplace=True)

player = pd.merge(batting, bowling,on=["match_id", "player"],how="outer").fillna(0)

player.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(player["strike_rate"], player["run_scored"], alpha=0.6, color="green")
plt.title("Strike Rate vs Runs Scored")
plt.xlabel("Strike Rate")
plt.ylabel("Runs Scored")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(player["economy_rate"], player["wickets"], alpha=0.6, color="red")
plt.title("Economy Rate vs Wickets Taken")
plt.xlabel("Economy Rate")
plt.ylabel("Wickets")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 run scorers overall
top_batsmen = player.groupby("player")["run_scored"].sum().sort_values(ascending=False).head(10)
#  Yeh har player ke naam ke hisaab se groups banata hai. Ek player ke sabhi matches ke record
#  ["runs_scored"] = keep the data of one column i.e, runs_scored
# sum() = har player ke matches ke runs ka sum
# .sort_values(ascending=False) = Total runs ko descending order me arrange kar rahein hai.


plt.figure(figsize=(12,6))
top_batsmen.plot(kind="bar", color="skyblue")
plt.title("Top 10 Run Scorers")
plt.ylabel("Total Runs")
plt.xlabel("Player")
plt.show()
```

```python
# Total wickets by player
top_bowlers = player.groupby("player")["wickets"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_bowlers.plot(kind="bar", color="orange")
plt.title("Top 10 Wicket Takers")
plt.ylabel("Wickets Taken")
plt.xlabel("Player")
plt.show()
```

```python
player["good_batting"] = player["run_scored"].apply(lambda x: 1 if x >= 30 else 0)
player["good_bowling"] = player["wickets"].apply(lambda x: 1 if x >= 2 else 0)

## agar 30 se jyada runs hai to good_batting = 1 ... ek match ke score
# agar 2 se jyada wicket li to good bowling ... ek match ke wickets
```

```python
# Features for predicting good batting
features_batting = ["ball_faced", "strike_rate", "wickets", "economy_rate"]  # model ko input diya 
label_batting = "good_batting"  ## usko good batting ka label denge jab 30 + score hoga

# Matlab model ko ye sikhana hai: “given features → predict good batting (1/0)”.
```

```python
# Features for predicting good bowling
features_bowling = ["run_scored", "ball_bowled", "over", "economy_rate"]  
label_bowling = "good_bowling"

# good bowling when wicket is 2+
#  "good_bowling" → 1 agar usne ≥ 2 wickets liye.
# "good_bowling" → 0 agar usne < 2 wickets liye.
```

```python
from sklearn.model_selection import train_test_split

# Batting
X_bat = player[features_batting]
y_bat = player[label_batting]

X_train_bat, X_test_bat, y_train_bat, y_test_bat = train_test_split(X_bat, y_bat, test_size=0.2, random_state=42)

# Bowling
X_bowl = player[features_bowling]
y_bowl = player[label_bowling]

X_train_bowl, X_test_bowl, y_train_bowl, y_test_bowl = train_test_split(X_bowl, y_bowl, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Batting model
bat_model = RandomForestClassifier(n_estimators=100, random_state=42)
# n_estimators=100 ==> 100 decision tree banege

bat_model.fit(X_train_bat, y_train_bat)   #  Model ko batting training data se train kar diya.
y_pred_bat = bat_model.predict(X_test_bat)  # Ab trained model ne test batting data pe prediction kiya (0/1).

print("Batting Accuracy:", accuracy_score(y_test_bat, y_pred_bat))
print(classification_report(y_test_bat, y_pred_bat))

# Precision → model ne jitne ko “good batting” bola, unme se kitne actually good the.
# Recall → jitne actually “good batting” the, unme se kitne ko model ne pakda.
# F1-score → precision aur recall ka balance.

# Bowling model
bowl_model = RandomForestClassifier(n_estimators=100, random_state=42)
bowl_model.fit(X_train_bowl, y_train_bowl)
y_pred_bowl = bowl_model.predict(X_test_bowl)

# Model ko bowling data pe train kiya aur prediction nikal liya.


print("Bowling Accuracy:", accuracy_score(y_test_bowl, y_pred_bowl))
print(classification_report(y_test_bowl, y_pred_bowl))
```

## Result :
1.in batting result is accurate  and balanced matrices

-Precision 0.98 (class 1) → model ne jo "good batting" bola, unme 98% sahi the.
-Recall 0.95 (class 1) → actual good batting me se 95% ko model ne correctly identify kiya.
-F1-score 0.96 → precision aur recall ka balance.

2. bowling ( imbalanced dataset = 0 (not good) kaafi zyada tha, aur 1 (good) kam)

- Precision 0.48 → model ne 48% sahi identify kiya jo "good bowling" bola.
- Recall 0.31 → actual good bowling me se sirf 31% correctly detect hua.
- F1-score 0.38 → bahut low, matlab model minority class ko sahi identify nahi kar raha.


```python
# SMOTE = technique to handle imbalanced data
# counter = classes ka count check karne ke liye.

from imblearn.over_sampling import SMOTE
from collections import Counter


## before smote
# Matlab class 0 (not good bowling) zyada hai, aur class 1 (good bowling) kam hai.
print("Before SMOTE:", Counter(y_bowl))

#  Apply SMOTE (oversampling minority class)
smote = SMOTE(random_state=42)
X_bowl_res, y_bowl_res = smote.fit_resample(X_bowl, y_bowl)

#  Check new class distribution
print("After SMOTE:", Counter(y_bowl_res))

#  Split into train/test
X_train_bowl, X_test_bowl, y_train_bowl, y_test_bowl = train_test_split(X_bowl_res, y_bowl_res, test_size=0.2, random_state=42)

#  Train model again
bowl_model = RandomForestClassifier(n_estimators=100, random_state=42)
bowl_model.fit(X_train_bowl, y_train_bowl)
y_pred_bowl = bowl_model.predict(X_test_bowl)

#  Evaluate
print("Bowling Accuracy:", accuracy_score(y_test_bowl, y_pred_bowl))
print(classification_report(y_test_bowl, y_pred_bowl))



print("Batting Accuracy:", accuracy_score(y_test_bat, y_pred_bat))
print(classification_report(y_test_bat, y_pred_bat))
```

## Before SMOTE  =Counter({0: 817, 1: 221})
 -Class 0 (not good bowling) bahut zyada hai.
- Class 1 (good bowling) bahut kam → imbalanced dataset.

## After SMOTE = Counter({0: 817, 1: 817})
- Minority class (1) ko oversample karke equal kar diya.
-Dataset ab balanced hai → model dono classes pe equally focus karega.
